```python
import pickle
import numpy as np
```

```python
with open('train_acc.pickle', 'rb') as f:
    train_acc = pickle.load(f)
```

```python
train_acc = np.asarray(train_acc)
```

```python
train_acc.shape
```

```python
with open('valid_acc.pickle', 'rb') as f:
    valid_acc = pickle.load(f)
```

```python
valid_acc= np.asarray(valid_acc)
```

```python
valid_acc.shape
```

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
epochs =list(np.arange(1,11))
```

```python
plt.plot(epochs, train_acc, label='Train Acc.')
plt.plot(epochs, valid_acc, label='Valid Acc.')
```

```python
plt.plot(epochs, list(train_acc), label='Train Acc.')
plt.plot(epochs, list(valid_acc), label='Valid Acc.')
plt.title('Epoch - Train/Valid Acc.')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```
